Логика парсинга:
1) Парсим данные с sravni.ru

2) Парсим по 6 категориям: 

- вклад, кредит, автокредит, ипотека, кредитная карта, дебетовая карта

3) По каждой категории нас интерсует по 2000 отзывов с sravni

4) По отзыву будут спарсшены: 

- текст, метка положительный\негативный, название банка, дата, город (регион), количество звезд и возможно что-то ещё


```python
import re
import time
from collections import defaultdict
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from tqdm import tqdm
```

```python
def get_category(url):
    driver = webdriver.Safari()
    driver.get(url)
    time_finished = 0
    SCROLL_PAUSE_TIME = 0.4
    scroll_height = 200
    num_of_responses = 0
    progress_bar = tqdm(total=750)
    try:
        while time_finished <= 750:
            driver.execute_script("window.scrollTo(0, {});".format(scroll_height))
            scroll_height += 200
            time.sleep(SCROLL_PAUSE_TIME)
            time_finished += SCROLL_PAUSE_TIME
            if (time_finished + 0.01) % 10 < 0.1:
                progress_bar.update(10)
    except:
        print('skip')
    progress_bar.close()
    tree = BeautifulSoup(driver.page_source)
    main_part = tree.find('div', {'class' : "styles_wrapper___EM4q"})
    paragraphs = main_part.find_all('a', {'target': '_blank'}, recursive = False)
    print(len(paragraphs))
    # Раскрываю все текста
    counter = 0
    for child in paragraphs:
        counter += 1
        a = child.find('a', {'class': "h-ml-4 _p7lcln _7k5cz5"})
        if a != None:
            driver.find_element(By.XPATH, f'//*[@id="__next"]/div[4]/div[1]/div[1]/div[2]/div/a[{counter}]/div/div/div[2]/a').click()
            time.sleep(0.2)
    tree = BeautifulSoup(driver.page_source)
    main_part = tree.find('div', {'class' : "styles_wrapper___EM4q"})
    paragraphs = main_part.find_all('a', {'target': '_blank'}, recursive = False)


    driver.quit()
    return paragraphs

category_trees = {
 'savings': {}, 
 'credits': {}, 
 'creditCards': {},
 'mortgage': {},
 'autocredits': {},
 'debitCards': {}
}  


for category in tqdm(category_trees):
    paragraphs = get_category(f'https://www.sravni.ru/banki/otzyvy/?filterBy=all&tag={category}')
    category_trees[category] = paragraphs
```

```python
def get_text(response, category, number):
    texts = response.find('div', {
        'class': "review-card_text__jTUSq articleTypography_article-comment__Px4n0"}).find_all('p')
    text = []
    for par in texts:
        if par.name:
            temp = par.text.strip()
            if temp != '':
                text.append(temp)
    responses_by_category[category][number]['text'] = ' '.join(text).replace('\xa0', ' ')

def get_stars(response, category, number):
    try:
        path = response.find('div', {'data-qa' : 'Rate'})
        if path != None:
            stars = path.find_all('span', {'class' : '_n6sdlh _akgnn3 _8136q1 _1xowbbl _18zu2v3 _g18kgu shape-round'}) 
            num_of_stars = len(stars)
        else:
            num_of_stars = -1
        responses_by_category[category][number]['rating'] = num_of_stars 
        if num_of_stars == -1:
            responses_by_category[category][number]['grade'] = 'unknown'
        elif num_of_stars <= 2:
            responses_by_category[category][number]['grade'] = 'bad'
        else:
            responses_by_category[category][number]['grade'] = 'good'
    except:
        print("it's broken here")
        
def get_city(response, category, number):
    try:
        path = response.find('div', {'class' : 'h-color-D30 h-ml-8 _1h41p0x _1gpt55s'})
        if path.text and path.text.strip() != '':
            responses_by_category[category][number]['city'] = path.text.strip()          
    except:
        print("it's broken here")

def get_date(response, category, number):
    try:
        path = response.find('div', {'class' : "_1n8o0h2 _vea58f"}).find('div', {'class' : 'h-color-D30 _1h41p0x'})
        if path.text and path.text.strip() != '':
            responses_by_category[category][number]['date'] = path.text.strip()          
    except:
        print("it's broken here") 

def get_bank_name(response, category, number):
    try:
        path = response.find('div', {'class' :"h-color-D100 _1h41p0x"})
        if path.text and path.text.strip() != '': 
            responses_by_category[category][number]['bank'] = path.text.strip()
    except:
        print("it's broken here")        
```

```python
responses_by_category = {
 'savings': {}, 
 'credits': {}, 
 'creditCards': {},
 'mortgage': {},
 'autocredits': {},
 'debitCards': {}
}  

for category in tqdm(responses_by_category):
    responses = category_trees[category]
    for response in tqdm(responses):
        number = response.get('href')[-7:-1]
        responses_by_category[category][number] = {
                                                  'text' : '',
                                                  'city' : '',
                                                  'date' : '',
                                                  'rating' : 0,
                                                  'grade' : '',
                                                  'bank' : ''
                                                  } 
        get_stars(response, category, number)
        get_date(response, category, number)
        get_city(response, category, number)
        get_text(response, category, number)
        get_bank_name(response, category, number) 
        
        
```
