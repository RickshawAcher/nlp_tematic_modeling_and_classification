```python
from classes_and_definitions import *
from datetime import datetime
import pickle
```

```python
with open('./Вспомогательные файлы/banki_ru.json') as f:
    banki_ru = json.load(f)
with open('./Вспомогательные файлы/sravni_ru.json') as f:
    sravni_ru = json.load(f)
    
df = pd.DataFrame(banki_ru['credits']).T
df['category'] = 'credits'
df['site'] = 'banki_ru'
len_of_frame = len(df)
for category in banki_ru:
    if category != 'credits':
        temp = pd.DataFrame(banki_ru[category]).T
        len_of_frame += len(temp)
        temp['category'] = category
        temp['site'] = 'banki_ru'
        df = pd.concat([df, temp], axis = 0)
    
for category in sravni_ru:
    temp = pd.DataFrame(sravni_ru[category]).T
    len_of_frame += len(temp)
    temp['category'] = category
    temp['site'] = 'sravni_ru'
    df = pd.concat([df, temp], axis = 0)
print('Должно быть:', len_of_frame)
print('Выдаёт:', len(df))

for column in df.columns:
    if column == 'rating':
        continue
    df = df[df[column].apply(lambda x: x.strip() != '')]

df.head(3)
```

```python
date_mapping = {'января': '01', 'февраля': '02', 'марта': '03',
'апреля': '04', 'мая': '05', 'июня': '06', 
 'июля': '07', 'августа': '08', 'сентября': '09',
 'октября': '10', 'ноября': '11', 'декабря': '12'}

def process_city(x):
    left = x.find('(')
    if left == -1:
        return x
    return x[:left].strip()

def process_date_string(x, date_mapping = date_mapping):
    today = datetime.today()
    try:
        if '.' in x:  # Если формат '16.01.2024 20:32'
            return x.split()[0]
        elif ':' in x:  # Если формат '13:18'
            return '16.02.2024'
        elif '2023' in x:  # Если формат '16 октября 2023'
            x_processed = x.split()[:2]
            x_processed[1] = date_mapping[x_processed[1]]
            if len(x_processed[0]) == 1:
                x_processed[0] = '0' + x_processed[0]
            return '.'.join(x_processed) + '.' + '2023'
        else:  # Если формат '13 февраля'
            x_processed = x.split()
            if len(x_processed[0]) == 1:
                x_processed[0] = '0' + x_processed[0]
            x_processed[1] = date_mapping[x_processed[1]]
            return '.'.join(x_processed) + '.' + '2024'
    except:
        print(x)

def translate_categories(x):
    if x == 'savings':
        return 'deposits'
    elif x == 'creditCards':
        return 'creditcards'
    elif x == 'debitCards':
        return 'debitcards'
    elif x =='mortgage':
        return 'hypothec'
    else:
        return x

def correctize_bank_name(x):
    '''
    Привожу к формату banki.ru
    '''
    if x == 'Кредит Европа Банк (Россия)':
        return 'Кредит Европа Банк'
    elif x == 'Банк БЖФ':
        return 'БЖФ Банк'
    elif x == 'Росбанк Дом':
        return 'Росбанк'
    elif x == 'Центр-инвест':
        return 'Банк «Центр-инвест»'
    elif x == 'Азиатско-Тихоокеанский Банк':
        return 'Азиатско-Тихоокеанский банк (АТБ)'
    elif x == 'Райффайзенбанк':
        return 'Райффайзен Банк'
    elif x == 'Россия':
        return 'Россия'
    elif x == 'МТС-Банк':
        return 'МТС Банк'
    elif x == 'Камкомбанк':
        return 'КАМКОМБАНК'
    elif x == 'Уральский Банк реконструкции и развития':
        return 'Уральский банк реконструкции и развития (УБРиР)'
    elif x == 'ОЗОН Банк':
        return 'Ozon Банк'
    elif x == 'Руснарбанк':
        return '«РУСНАРБАНК»'
    elif x == 'РОСБАНК':
        return 'Росбанк'
    elif x == 'Кубань Кредит':
        return '«Кубань Кредит»'
    elif x == 'МСП Банк':
        return 'СМП Банк'
    elif x == 'ФК Открытие':
        return 'Банк «Открытие»'
    elif x == 'Русский стандарт':
        return 'Русский Стандарт'
    elif x == 'Банк Уралсиб':
        return 'Уралсиб'
    elif x == 'Россия':
        return 'Банк «РОССИЯ»'
    else:
        return x

df['city'] = df['city'].apply(process_city)
df['date'] = pd.to_datetime(df['date'].apply(process_date_string), format='%d.%m.%Y')
df['week'] = df['date'].dt.to_period('W')
df['month'] = df['date'].dt.to_period('M')  
df['category'] = df['category'].apply(translate_categories)
df['bank'] = df['bank'].apply(correctize_bank_name)
```
